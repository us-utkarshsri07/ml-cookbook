```python
#6.1 Cleaning Text 

text_data=[" Interrobang. by Aishwarya Henritte","Parking And Going. by Karl Gautier","Today Is The night. By Jarek Prakash"]
strip_whitespace=[s.strip() for s in text_data]
strip_whitespace
```

```python
#Remove periods
remove_periods=[s.replace(".","") for s in strip_whitespace]
remove_periods
```

```python
def capitalizer(string:str)-> str:
    return string.upper()
capitalized=[capitalizer(s) for s in remove_periods]
```

```python
capitalized
```

```python
import re
def replace_letters_with_X(string:str)-> str:
    return re.sub("[a-zA-Z]","X",string)
replaced=[replace_letters_with_X(s) for s in capitalized]
replaced
```

```python
#6.3 Removing Punctuation
import unicodedata
import sys
text_data=['hi!!! I. Love. This. Song.....','10000% Agree!!!! #LoveIT','Right?!!!!']
punctuation=dict.fromkeys(i for i in range(sys.maxunicode) if unicodedata.category(chr(i)).startswith('P')) 
[string.translate(punctuation) for string in text_data]
```

```python
#6.5 Removing Stop Words
from nltk.corpus import stopwords
import nltk
nltk.download('stopword')
tokenize_word=['i', 'love', 'this', 'song'  
            ,'10000', 'agree', 'loveit', 'right']
stopwords=set(stopwords.words('english'))
[word for word in tokenize_word if word not in stopwords]
```

```python
#6.6 Stemming Words
from nltk.stem import PorterStemmer
tokenize_word=['running', 'ran', 'runs', 'easily', 'fairly']
ps=PorterStemmer()
[ps.stem(word) for word in tokenize_word]
```

```python
#6.7 Tagging Parts of Speech
from nltk import pos_tag
from nltk.tokenize import word_tokenize
text_data="The science of today is the technology of tomorrow"
text_tagged=pos_tag(word_tokenize(text_data))
text_tagged
```

```python
'''TAG 
NNP-proper noun,singular,
NN-Noun,singular or mass,
RB-Adverb,
VBD-Verb,past tense,
VBG-Verb,gerund or present participle,
JJ-Adjective,
PRP-Personal pronoun'''
```

```python
#6.8 Encoding Text as a Bag of Words
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
text_data=np.array(["I love this song","I love it","I really love this song"])
count=CountVectorizer()
bag_of_words=count.fit_transform(text_data)
bag_of_words
```

```python
bag_of_words.toarray()
```

```python
count.get_feature_names_out()
```

```python
count_2gram=CountVectorizer(ngram_range=(2,2),stop_words='english',vocabulary=['brazil'])
bag=count_2gram.fit_transform(text_data)
bag.toarray()
```

```python
count_2gram.vocabulary_
```

```python
#6.9 Weighting Word Importance 
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
text_data=np.array(["I love this song","I love it","I really love this song"])
tfidf=TfidfVectorizer()
feature_matrix=tfidf.fit_transform(text_data)
feature_matrix
```

```python
feature_matrix.toarray()
```
