```python
#1.3 Creating a Sparse Matrix(non-zero element)
from scipy import sparse
matrix=np.array([[0,0],[0,1],[3,0]])
#Create a compressed sparse row matrix
matrix_sparse=sparse.csr_matrix(matrix)
print(matrix_sparse)
```

```python
#1.17 calculating the dot products
vec_a=np.array([1,2,3])
vec_b=np.array([4,5,6])
np.dot(vec_a,vec_b)
```

```python
#1.18 Adding and Subtracting Matrices
matrix_a=np.array([[1,1,1],[1,1,1],[1,1,2]])
matrix_b=np.array([[1,3,1],[1,3,1],[1,3,8]])
np.add(matrix_a,matrix_b)
```

```python
np.subtract(matrix_a,matrix_b)
```

```python
#Alternate
matrix_a + matrix_b
```

```python
#1.1 Multiplying Matrices
mat_a=np.array([[1,1],[1,2]])
mat_b=np.array([[1,3],[1,2]])
np.dot(mat_a,mat_b)
```

```python
mat_a @ mat_b
```

```python
mat_a * mat_b
```

```python
# 1.20 Inverting a matrix 
# AA^-1=I
matrix=np.array([[1,4],[2,5]])
np.linalg.inv(matrix)
```

```python
matrix @ np.linalg.inv(matrix)
```
