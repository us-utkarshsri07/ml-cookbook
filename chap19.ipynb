```python
#19.1 Clustering Using K-Means

from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

iris = datasets.load_iris()
X = iris.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
kmeans = KMeans(n_clusters=3, random_state=42)
model=kmeans.fit(X_scaled)
```

```python
model.labels_
```

```python
#19.2 Speeding up k-means clustering 
from sklearn import datasets
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import StandardScaler

iris = datasets.load_iris()
X = iris.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
kmeans = MiniBatchKMeans(n_clusters=3, batch_size=3072, random_state=42)
model=kmeans.fit(X_scaled)
```

```python
#19.3 Clustering Using Mean Shift
from sklearn import datasets
from sklearn.cluster import MeanShift
from sklearn.preprocessing import StandardScaler

iris = datasets.load_iris()
X = iris.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)



meanshift = MeanShift(bandwidth=1.5)
model=meanshift.fit(X_scaled)       
```

```python
#19.4 Clustering Using DBSCAN
from sklearn import datasets
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
iris = datasets.load_iris()
X = iris.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
dbscan = DBSCAN(eps=0.5, min_samples=5)
model=dbscan.fit(X_scaled)
```

```python
model.labels_
```

```python
#19.5 Clsutering Using Hierarchical Merging
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

iris = datasets.load_iris()
X = iris.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

agglo = AgglomerativeClustering(n_clusters=3, linkage='ward')

model=agglo.fit(X_scaled)
```

```python
model.labels_
```
