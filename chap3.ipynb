```python
#3.14 Grouping Rows by Time
import numpy as np
import pandas as pd
time_index=pd.date_range('20230101',periods=6)
dataframe=pd.DataFrame({'A':np.random.randn(6),'B':np.random.randn(6)},index=time_index)
dataframe['Sales_Amount']=np.random.randint(1,10,6)
dataframe.resample('W').sum()
```

```python
#3.15 Looping over a Column
df=pd.read_csv('tested.csv')
```

```python
# 3.16 Applying A function over all elements in a column
def uppercase(name):
    return name.upper()
df['Name'].apply(uppercase)[0:2] #show two rows
```

```python
#3.17 Applying a Function to groups
df.groupby('Sex').apply(lambda name: name.count())
```

```python
# 3.18 Concatenating DataFrames
#Use concat with axis=0 to concatenate rows
data_a = {'Name': ['John', 'Jane', 'Doe'],
          'Age': [28, 34, 29]  }
data_b = {'Name': ['Alice', 'Bob', 'Charlie'],
          'Age': [25, 30, 35]  }
df_a = pd.DataFrame(data_a)
df_b = pd.DataFrame(data_b)
pd.concat([df_a, df_b], axis=0, ignore_index=True)
```

```python
pd.concat([df_a, df_b])
```

```python
#3.19 Merging DataFrames
data_a = {'Name': ['John', 'Jane', 'Doe'],
            'Age': [28, 34, 29]  }  
data_b = {'Name': ['Jane', 'Doe', 'Alice'],
            'City': ['Los Angeles', 'Chicago', 'San Francisco']  }
df_a = pd.DataFrame(data_a)
df_b = pd.DataFrame(data_b)
pd.merge(df_a, df_b, on='Name', how='outer') #how='inner','outer','left','right'
```
